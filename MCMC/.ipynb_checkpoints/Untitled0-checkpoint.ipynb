```python
#     Importamos los datos

data = loadtxt("movimiento.dat")
x_data = data.T[0]
y_data = data.T[1]

scatter(x_data, y_data)
```

```python
#     Para corregir el error, debemos quitar la exponencial del return

def likelihood(y_obs, y_model):
    chi_sq = (0.5)*sum( (y_obs-y_model)**2 )
    return -chi_sq
```

```python
#     Definimos nuestro modelo

def model(t_obs, v0, g):
    return v0*t_obs+( 0.5*g*( t_obs**2 ) )
```

```python
#     Creare mis propios parametros de avance con juegos de azar y mujerzuelas

v0_walk = []
g_walk = []
l_walk = []

v0_walk = append(v0_walk, random.random())
g_walk = append(g_walk, random.random())

init_pos = model(x_data, v0_walk[0], g_walk[0])
l_walk = append(l_walk, likelihood(y_data, init_pos))

print v0_walk, g_walk, l_walk
```

```python
#     Mecanismo de la caminata

n_iterations = 20000

for i in range(n_iterations):
    v0_prime = np.random.normal(v0_walk[i], 0.1) 
    g_prime = np.random.normal(g_walk[i], 0.1)

    y_init = model(x_data, v0_walk[i], g_walk[i])
    y_prime = model(x_data, v0_prime, g_prime)
    
    l_prime = likelihood(y_data, y_prime)
    l_init = likelihood(y_data, y_init)
    
    alpha = l_prime/l_init
    if(alpha<1.0):
        v0_walk  = append(v0_walk,v0_prime)
        g_walk  = append(g_walk,g_prime)
        l_walk = append(l_walk, l_prime)
    else:
        beta = random.random()
        if(beta<=exp(-alpha)):
            v0_walk = append(v0_walk,v0_prime)
            g_walk = append(g_walk,g_prime)
            l_walk = append(l_walk, l_prime)
        else:
            v0_walk = append(v0_walk,v0_walk[i])
            g_walk = append(g_walk,g_walk[i])
            l_walk = append(l_walk, l_init)
```

```python
# Visualizacion de el avance

scatter(v0_walk, g_walk)
```

```python
#no tengo que hacer log porque quite la exponencial

scatter(v0_walk, -(l_walk))
```

```python
#     mejor g

count, bins, ignored =plt.hist(g_walk, 50, normed=True)
```

```python
#     extraccion de los parametros optimos

max_likelihood_id = argmax(l_walk)
best_v0 = v0_walk[max_likelihood_id]
best_g = g_walk[max_likelihood_id]
print l_walk[max_likelihood_id]
print best_v0
print best_g
```

```python
#     Visualizacion final

best_y = model(x_data, best_v0, best_g)
scatter(x_data,y_data)
plot(x_data, best_y)
```
