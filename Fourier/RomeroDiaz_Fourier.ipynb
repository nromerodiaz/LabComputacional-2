# Taller: Transformada de Fourier

```python
%pylab inline
from scipy.fftpack import fft, fftfreq
```

```python
#     Creamos nuestros puntos aleatorios

n_points = 1000
func = random.random(n_points)

#     Restando el promedio:

func = func - np.mean(func)
```

$\LARGE\mathbf{\cdot}$ Graficando la señal generada:

```python
plot(func)
```

$\LARGE\mathbf{\cdot}$Ahora debemos tratar esta señal con la transformada de Fourier

```python
#     Aplicando la transformada

F_func = fft(func) / n_points
freq = fftfreq(n_points, 1)
```

$\LARGE\mathbf{\cdot}$ Graficando:

```python
plot(freq, abs(F_func))
```

```python
n_half = ceil(n_points/2)
F_func = 2*F_func[:n_half]
freq_2 = freq[:n_half]
scatter(freq_2, log10(np.absolute(F_func)**2)) 

xlim([-0.09, 0.6])
ylim([-7, -1])
```

```python
#     Normalizando:

F_norm = F_func/np.absolute(F_func)
```

$\LARGE\mathbf{\cdot}$ Graficando:

```python
plot(F_norm-1)
```

```python
#     Aplicando la transformada inversa:

new_F = ifft(F_norm)
```

$\LARGE\mathbf{\cdot}$ Graficando:

```python
#     Este es el ruido

plot(new_F)
```

Ahora, podemos comparar el ruido de fondo con nuestra señal original:

```python
plot(new_F)
plot(F_norm, alpha= 0.4, c ='r')
```
